```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
centers= [[5,5],[-3,-3],[7,-9]]
cluster_std= [1,1,1]
x,y= make_blobs(n_samples=500, centers=centers, cluster_std=cluster_std,n_features=2, random_state=0)
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
wcss= []
for i in range(1,11):
    kmeans= KMeans(n_clusters=i, init='k-means++',n_init='auto', random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
```

```python
wcss
```

```python
# SO the elbow point is 3
km= KMeans(n_clusters=3, init='k-means++',n_init='auto', random_state=0)
km.fit_predict(x)
y_pred=km.labels_
```

```python
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],c='red',label='Cluster 1')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],c='blue',label='Cluster 2')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],c='green',label='Cluster 3')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c='yellow',label='Centroids')
plt.legend()
```
